## Classificador autom√°tico de sentimento


A empresa que estamos avaliando √© a Nike, que fabrica materiais esportivos. Com um vi√©s de marketing, decidimos que √© interessante para a empresa saber como est√° sua reputa√ß√£o no mercado, portanto, se est√£o falando bem ou mal dela. Por isso, classificamos os tweets entre Negativos, Indiferentes/Irrelevantes (Neutros) ou Positivos.

Os positivos s√£o aqueles tweets que falam bem e constituem em sentimentos positivos com rela√ß√£o √† empresa, enquanto os negativos apresentam cr√≠ticas e sentimentos negativos. Os neutros s√£o os tweets que falam de outros assuntos ou apresentam sentimentos indiferentes.

___
### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.


$P(Positivo|Tweet) = \frac{P(Tweet|Positivo).P(Positivo)}{P(Tweet)}$


$P(Neutro|Tweet) = \frac{P(Tweet|Neutro).P(Neutro)}{P(Tweet)}$


$P(Negativo|Tweet) = \frac{P(Tweet|Negativo).P(Negativo)}{P(Tweet)}$


```python
# Probabilidade dos tweets Treinamento dado classifica√ß√£o

P1_Neutro = 0.412
P1_Negativo = 0.306
P1_Positivo = 0.282

Œ± = 1 # Valor estimado
V = 10e5 # Valor estimado
```

#### Para compara√ß√£o entre as probabilidades:

$Produto Positivo = \frac{P(Tweet|Positivo).P(Positivo)}{1}$

$Produto Neutro = \frac{P(Tweet|Neutro).P(Neutro)}{1}$

$Produto Negativo = \frac{P(Tweet|Negativo).P(Negativo)}{1}$

#### Smoothing:

$(Tweet|Positivo) = \frac{Contagem Positivo + Œ±}{Contagem Total + Œ±V}$

$(Tweet|Neutro) = \frac{Contagem Neutro + Œ±}{Contagem total + Œ±V}$

$(Tweet|Negativo) = \frac{Contagem Negativo + Œ±}{Contagem total + Œ±V}$

```python
i = 0 #contador treinamento

train["Classifica√ß√£o Naive-Bayes"] = "Batata"

for tweet in train['Treinamento']:
    
    # Limpando tweets do treinamento
    T_limpo = cleanup(tweet.lower())
    T_lista = pd.Series(T_limpo.split())
    
    # Fazendo o c√°lculo do Produto em Positivo
    P1_tweet_dado_pos = 1
    for palavra in T_lista:
        if palavra in todas_palavras_abs_pos:
            P1_palavra_dado_pos = (todas_palavras_abs_pos[palavra] + Œ±)/(todas_palavras_abs_pos.sum() + Œ±*V)
        else:
            P1_palavra_dado_pos = (Œ±)/(todas_palavras_abs_pos.sum() + Œ±*V)
        P1_tweet_dado_pos *= P1_palavra_dado_pos
    P1_produto_pos = P1_tweet_dado_pos * P1_Positivo
        
    # Fazendo o c√°lculo do Produto em Neutro
    P1_tweet_dado_neutro = 1
    for palavra in T_lista:
        if palavra in todas_palavras_abs_neutro:
            P1_palavra_dado_neutro = (todas_palavras_abs_neutro[palavra] + Œ±)/(todas_palavras_abs_neutro.sum() + Œ±*V)
        else:
            P1_palavra_dado_neutro = (Œ±)/(todas_palavras_abs_neutro.sum() + Œ±*V)
        P1_tweet_dado_neutro *= P1_palavra_dado_neutro
    P1_produto_neutro = P1_tweet_dado_neutro * P1_Neutro
        
    # Fazendo o c√°lculo do Produto em Negativo
    P1_tweet_dado_neg = 1
    for palavra in T_lista:
        if palavra in todas_palavras_abs_neg:
            P1_palavra_dado_neg = (todas_palavras_abs_neg[palavra] + Œ±)/(todas_palavras_abs_neg.sum() + Œ±*V)
        else:
            P1_palavra_dado_neg = (Œ±)/(todas_palavras_abs_neg.sum() + Œ±*V)
        P1_tweet_dado_neg *= P1_palavra_dado_neg
    P1_produto_neg = P1_tweet_dado_neg * P1_Negativo  
    
    #Comparando para identificar a categoria do tweet
    
    if P1_produto_pos > P1_produto_neg and P1_produto_pos > P1_produto_neutro:
        train["Classifica√ß√£o Naive-Bayes"].iloc[i] = "Positivo"

    if P1_produto_neutro > P1_produto_pos and P1_produto_neutro > P1_produto_neg:
        train["Classifica√ß√£o Naive-Bayes"].iloc[i] = "Neutro"
    
    if P1_produto_neg > P1_produto_neutro and P1_produto_neg > P1_produto_pos:
        train["Classifica√ß√£o Naive-Bayes"].iloc[i] = "Negativo"
    
    i+=1
    
train.head(30)
```

##### Se o Produto Positivo for maior que o Produto Negativo e o Produto Neutro, o classificador Naive-Bayes indica que esse tweet √© positivo.

##### Se o Produto Neutro for maior que o Produto Positivo e o  Produto Negativo , o classificador Naive-Bayes indica que esse tweet √© neutro.

##### Se o Produto Negativo for maior que o Produto Neutro e o Produto Positivo, o classificador Naive-Bayes indica que esse tweet √© negativo.

-----
Isso ocorre porque, ao calcular a desigualdade, √© poss√≠vel "cortar" o denominador da f√≥rmula do teorema de Bayes, tornando a compara√ß√£o independente desse denominador.

<p> No momento de classifica√ß√£o manual dos tweets, as mensagens que foram notadas como dupla nega√ß√£o e sarcasmo foram identificadas de acordo com a sua inten√ß√£o, e n√£o de acordo com o significado objetivo dos termos. Isso se d√° pois ele considera as palavras indepententes da ordem e da sua sequ√™ncia e, com isso, o classificador √© incapaz de distinguir o significado real de uma palavra de sua inten√ß√£o. <p> Al√©m disso, devido a essa mesma limita√ß√£o, ele tamb√©m √© impreciso ao identificar frases que contenham refer√™ncias internas a grupos reduzidos de pessoas, como por exemplo express√µes que indicam memes ou piadas, e que, nesse caso, tamb√©m distorcem o significado puro da palavra.</p>

<p> Assim, sem considerar a ordem e a interdepend√™ncia entre as palavras, √© imposs√≠vel para o computador identificar e distinguir a ocorr√™ncia de dupla-nega√ß√£o e de palavras com significado alterado, sarc√°stico ou quando s√£o acompanhadas de uma express√£o que indica ironia.</p>
