## Classificador automático de sentimento


A empresa que estamos avaliando é a Nike, que fabrica materiais esportivos. Com um viés de marketing, decidimos que é interessante para a empresa saber como está sua reputação no mercado, portanto, se estão falando bem ou mal dela. Por isso, classificamos os tweets entre Negativos, Indiferentes/Irrelevantes (Neutros) ou Positivos.

Os positivos são aqueles tweets que falam bem e constituem em sentimentos positivos com relação à empresa, enquanto os negativos apresentam críticas e sentimentos negativos. Os neutros são os tweets que falam de outros assuntos ou apresentam sentimentos indiferentes.

___
### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.


$P(Positivo|Tweet) = \frac{P(Tweet|Positivo).P(Positivo)}{P(Tweet)}$


$P(Neutro|Tweet) = \frac{P(Tweet|Neutro).P(Neutro)}{P(Tweet)}$


$P(Negativo|Tweet) = \frac{P(Tweet|Negativo).P(Negativo)}{P(Tweet)}$


```python
# Probabilidade dos tweets Treinamento dado classificação

P1_Neutro = 0.412
P1_Negativo = 0.306
P1_Positivo = 0.282

α = 1 # Valor estimado
V = 10e5 # Valor estimado
```

#### Para comparação entre as probabilidades:

$Produto Positivo = \frac{P(Tweet|Positivo).P(Positivo)}{1}$

$Produto Neutro = \frac{P(Tweet|Neutro).P(Neutro)}{1}$

$Produto Negativo = \frac{P(Tweet|Negativo).P(Negativo)}{1}$

#### Smoothing:

$(Tweet|Positivo) = \frac{Contagem Positivo + α}{Contagem Total + αV}$

$(Tweet|Neutro) = \frac{Contagem Neutro + α}{Contagem total + αV}$

$(Tweet|Negativo) = \frac{Contagem Negativo + α}{Contagem total + αV}$

```python
i = 0 #contador treinamento

train["Classificação Naive-Bayes"] = "Batata"

for tweet in train['Treinamento']:
    
    # Limpando tweets do treinamento
    T_limpo = cleanup(tweet.lower())
    T_lista = pd.Series(T_limpo.split())
    
    # Fazendo o cálculo do Produto em Positivo
    P1_tweet_dado_pos = 1
    for palavra in T_lista:
        if palavra in todas_palavras_abs_pos:
            P1_palavra_dado_pos = (todas_palavras_abs_pos[palavra] + α)/(todas_palavras_abs_pos.sum() + α*V)
        else:
            P1_palavra_dado_pos = (α)/(todas_palavras_abs_pos.sum() + α*V)
        P1_tweet_dado_pos *= P1_palavra_dado_pos
    P1_produto_pos = P1_tweet_dado_pos * P1_Positivo
        
    # Fazendo o cálculo do Produto em Neutro
    P1_tweet_dado_neutro = 1
    for palavra in T_lista:
        if palavra in todas_palavras_abs_neutro:
            P1_palavra_dado_neutro = (todas_palavras_abs_neutro[palavra] + α)/(todas_palavras_abs_neutro.sum() + α*V)
        else:
            P1_palavra_dado_neutro = (α)/(todas_palavras_abs_neutro.sum() + α*V)
        P1_tweet_dado_neutro *= P1_palavra_dado_neutro
    P1_produto_neutro = P1_tweet_dado_neutro * P1_Neutro
        
    # Fazendo o cálculo do Produto em Negativo
    P1_tweet_dado_neg = 1
    for palavra in T_lista:
        if palavra in todas_palavras_abs_neg:
            P1_palavra_dado_neg = (todas_palavras_abs_neg[palavra] + α)/(todas_palavras_abs_neg.sum() + α*V)
        else:
            P1_palavra_dado_neg = (α)/(todas_palavras_abs_neg.sum() + α*V)
        P1_tweet_dado_neg *= P1_palavra_dado_neg
    P1_produto_neg = P1_tweet_dado_neg * P1_Negativo  
    
    #Comparando para identificar a categoria do tweet
    
    if P1_produto_pos > P1_produto_neg and P1_produto_pos > P1_produto_neutro:
        train["Classificação Naive-Bayes"].iloc[i] = "Positivo"

    if P1_produto_neutro > P1_produto_pos and P1_produto_neutro > P1_produto_neg:
        train["Classificação Naive-Bayes"].iloc[i] = "Neutro"
    
    if P1_produto_neg > P1_produto_neutro and P1_produto_neg > P1_produto_pos:
        train["Classificação Naive-Bayes"].iloc[i] = "Negativo"
    
    i+=1
    
train.head(30)
```

##### Se o Produto Positivo for maior que o Produto Negativo e o Produto Neutro, o classificador Naive-Bayes indica que esse tweet é positivo.

##### Se o Produto Neutro for maior que o Produto Positivo e o  Produto Negativo , o classificador Naive-Bayes indica que esse tweet é neutro.

##### Se o Produto Negativo for maior que o Produto Neutro e o Produto Positivo, o classificador Naive-Bayes indica que esse tweet é negativo.

-----
Isso ocorre porque, ao calcular a desigualdade, é possível "cortar" o denominador da fórmula do teorema de Bayes, tornando a comparação independente desse denominador.

##### Algumas considerações sobre dupla-negação e sarcasmo:

<p> No momento de classificação manual dos tweets, as mensagens que foram notadas como dupla negação e sarcasmo foram identificadas de acordo com a sua intenção, e não de acordo com o significado objetivo dos termos. Esse fato constitui um problema para classificador Naive-Bayes. Isso se dá pois ele considera as palavras indepententes da ordem e da sua sequência e, com isso, o classificador é incapaz de distinguir o significado real de uma palavra de sua intenção. Dessa maneira, o Naive-Bayes se torna ineficiente contra ironias, duplos sentidos e dupla-negação. </p>
    
<p> Além disso, devido a essa mesma limitação, ele também é impreciso ao identificar frases que contenham referências internas a grupos reduzidos de pessoas, como por exemplo expressões que indicam memes ou piadas, e que, nesse caso, também distorcem o significado puro da palavra.</p>

<p> Assim, sem considerar a ordem e a interdependência entre as palavras, é impossível para o computador identificar e distinguir a ocorrência de dupla-negação e de palavras com significado alterado, sarcástico ou quando são acompanhadas de uma expressão que indica ironia.</p>
